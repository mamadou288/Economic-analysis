```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Charger le fichier CSV
file_path = "../data/HRDataset_v14.csv"  # Adapte le chemin si nécessaire
data = pd.read_csv(file_path)

# Aperçu des premières lignes
print("Aperçu des données :")
display(data.head())
```

- Le nombre de valeurs non nulles

Cela peut nous aide à comprendre la structure des données avant de passer à des analyses plus détaillées.


```python
# Résumé des colonnes
print("\nRésumé des colonnes :")
data.info()
```

```python
# Statistiques descriptives
print("\nStatistiques descriptives :")
display(data.describe(include='all'))
```

```python
# Vérification des valeurs manquantes
missing_values = data.isnull().sum()
print("Valeurs manquantes par colonne :")
display(missing_values)
```

Les 207 valeurs manquantes dans la colonne DateofTermination indiquent probablement que ces employés sont toujours en poste. Voici comment gérer cette situation et documenter le processus dans ton notebook.
Créer une nouvelle colonne binaire `CurrentlyEmployed` :
   - `1` si l'employé est toujours en poste.
   - `0` sinon.
   
Cela facilitera les analyses futures concernant l'emploi actuel des employés.

```python
# Remplacer les valeurs manquantes par "Still Employed"
data['DateofTermination'].fillna('Still Employed', inplace=True)

# Créer une nouvelle colonne binaire pour indiquer si l'employé est toujours en poste
data['CurrentlyEmployed'] = data['DateofTermination'].apply(lambda x: 1 if x == 'Still Employed' else 0)

# Vérification des modifications
print("Aperçu des données après gestion des valeurs manquantes :")
display(data[['DateofTermination', 'CurrentlyEmployed']].head())
```

```python
# Calcul du pourcentage d'employés encore en poste
currently_employed_percentage = data['CurrentlyEmployed'].value_counts(normalize=True) * 100

# Affichage des résultats
print("Pourcentage d'employés actuellement en poste :")
display(currently_employed_percentage)
```

```python
# Création du pie chart
labels = ['Still Employed', 'Terminated']
plt.figure(figsize=(6, 6))
plt.pie(currently_employed_percentage, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Répartition des employés (Actifs vs Terminé)')
plt.show()
```

## Turnover par département

l'objectif est d'identifier les départements ayant le plus haut taux de turnover.  
Approche :
1. Grouper les données par département.
2. Calculer le pourcentage d'employés ayant quitté (CurrentlyEmployed = 0) pour chaque département.
3. Visualiser les résultats avec un diagramme en barres.

```python
# Calculer le turnover par département
turnover_by_department = data.groupby('Department')['CurrentlyEmployed'].apply(lambda x: (x == 0).mean() * 100).sort_values(ascending=False)

# Affichage des résultats
print("Turnover par département (en %):")
display(turnover_by_department)
```

```python
plt.figure(figsize=(10, 6))
turnover_by_department.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Turnover par département (en %)')
plt.ylabel('Pourcentage de turnover')
plt.xlabel('Département')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Le turnover varie significativement selon les départements. Voici les observations principales :

Production
Le turnover est le plus élevé, ce qui pourrait indiquer :
Des conditions de travail difficiles.
Un manque de satisfaction ou d'engagement parmi les employés.
Software Engineering
Un turnover élevé peut refléter :
Une forte demande sur le marché, attirant les talents vers d'autres entreprises.
Des problèmes internes tels que :
Une charge de travail excessive.
Un manque de reconnaissance ou de soutien.
Sales et Executive Office
Ces départements montrent un turnover plus faible, ce qui peut être le signe :
D'une meilleure rétention des employés.
De rôles plus stables ou attractifs.


Identifier les facteurs qui pourraient influencer le turnover, en analysant les corrélations entre :

- Satisfaction des employés (EmpSatisfaction)
- Score d'engagement (EngagementSurvey)
- Nombre de projets spéciaux (SpecialProjectsCount)
- Retards des 30 derniers jours (DaysLateLast30)



```python
# Variables à analyser
variables = ['EmpSatisfaction', 'EngagementSurvey', 'SpecialProjectsCount', 'DaysLateLast30', 'CurrentlyEmployed']

# Calcul de la matrice de corrélation
correlation_matrix = data[variables].corr()

# Visualisation de la heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation entre les variables")
plt.show()
```

Satisfaction des employés (EmpSatisfaction)
- Corrélation faible avec le turnover (CurrentlyEmployed = 0.00).
- La satisfaction des employés ne semble pas être un facteur direct influençant le turnover.

Score d'engagement (EngagementSurvey)
- Corrélation faible avec le turnover (0.02).
Une exploration plus détaillée, comme des scatter plots, pourrait révéler des tendances.

Nombre de projets spéciaux (SpecialProjectsCount)
- Corrélation positive modérée avec le turnover (0.15).
- Les employés impliqués dans plusieurs projets semblent plus susceptibles de quitter.

Retards des 30 derniers jours (DaysLateLast30)
- Corrélation négative avec le turnover (-0.14).
- Les employés encore en poste semblent être plus souvent en retard, ce qui pourrait indiquer un stress ou une surcharge.

Relations internes
- Forte corrélation négative entre l'engagement et les retards (-0.59).
- Les employés moins engagés ont tendance à être plus souvent en retard.

```python
# Colonnes à analyser pour les outliers
columns_to_check = ['EngagementSurvey', 'EmpSatisfaction', 'SpecialProjectsCount', 'DaysLateLast30']

outliers = {}

for col in columns_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identifier les outliers
    outliers[col] = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    print(f"Outliers pour la colonne {col} :")
    print(f"- Limite inférieure : {lower_bound}")
    print(f"- Limite supérieure : {upper_bound}")
    print(f"- Nombre d'outliers : {len(outliers[col])}\n")
    display(outliers[col])

# Visualisation des outliers pour une colonne
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=data[columns_to_check], orient="h", palette="Set2")
plt.title("Visualisation des outliers (règle des 1.5 * IQR)")
plt.show()
```

```python
# Comparer les performances pour les outliers dans SpecialProjectsCount
special_projects_outliers = data['SpecialProjectsCount'] > 0
plt.figure(figsize=(8, 6))
sns.boxplot(x=special_projects_outliers, y=data['PerformanceScore'], palette='Set3')
plt.title("Impact des outliers (SpecialProjectsCount) sur les performances")
plt.xlabel("Outliers (SpecialProjectsCount)")
plt.ylabel("PerformanceScore")
plt.show()

# Comparer les performances pour les outliers dans DaysLateLast30
days_late_outliers = data['DaysLateLast30'] > 0
plt.figure(figsize=(8, 6))
sns.boxplot(x=days_late_outliers, y=data['PerformanceScore'], palette='Set2')
plt.title("Impact des outliers (DaysLateLast30) sur les performances")
plt.xlabel("Outliers (DaysLateLast30)")
plt.ylabel("PerformanceScore")
plt.show()
```

Les employés identifiés comme outliers (avec un nombre de projets spéciaux supérieur à 0) semblent avoir des performances qui se répartissent principalement dans les catégories "Exceeds" et "Fully Meets".
Cela pourrait indiquer que les projets spéciaux sont souvent attribués aux employés les plus performants.
Les employés avec un nombre de projets spéciaux égal à 0 présentent des performances variées, couvrant toutes les catégories, y compris "Needs Improvement" et "PIP".

```python
# Retards par département
late_by_department = data.groupby('Department')['DaysLateLast30'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
late_by_department.plot(kind='bar', color='teal', edgecolor='black')
plt.title("Moyenne des jours de retard par département")
plt.xlabel("Département")
plt.ylabel("Moyenne des jours de retard")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

Les retards dans Sales peuvent indiquer des défis spécifiques au département, comme des déplacements ou une organisation différente des horaires.
Les départements avec peu de retards (comme Admin Offices) peuvent refléter une meilleure gestion du temps ou des processus bien établis.
Les retards peuvent être un indicateur de faible performance, surtout dans les cas où les scores tombent dans les catégories inférieures.
Cependant, il existe aussi des employés retardataires avec des scores élevés, ce qui suggère que d'autres facteurs influencent également la performance.

```python
# Boxplot des retards et des scores d'engagement
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['DaysLateLast30'] > 0, y=data['EngagementSurvey'], palette="muted")
plt.title("Relation entre les retards et le score d'engagement")
plt.xlabel("Employés avec retard (1=Oui, 0=Non)")
plt.ylabel("Score d'engagement")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Les employés avec des retards fréquents ont tendance à avoir des scores d'engagement plus faibles, ce qui peut indiquer un désengagement.
Cela suggère que les retards pourraient être un indicateur de perte de motivation ou de satisfaction au travail.

```python
# Boxplot des retards et des projets spéciaux
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['DaysLateLast30'] > 0, y=data['SpecialProjectsCount'], palette="coolwarm")
plt.title("Relation entre les retards et la surcharge de travail")
plt.xlabel("Employés avec retard (1=Oui, 0=Non)")
plt.ylabel("Nombre de projets spéciaux")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Il n'y a pas de preuve claire que la surcharge de travail (via le nombre de projets spéciaux) contribue aux retards.
Les retards pourraient être davantage liés à des facteurs comportementaux ou organisationnels, comme le désengagement (confirmé dans l'analyse précédente).

```python
# Boxplot : Retards et salaire
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['DaysLateLast30'] > 0, y=data['Salary'], palette="coolwarm")
plt.title("Relation entre les retards et le salaire")
plt.xlabel("Employés avec retard (1=Oui, 0=Non)")
plt.ylabel("Salaire")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Les employés retardataires semblent, en général, appartenir à une tranche de salaire inférieure.
Cela pourrait indiquer une corrélation entre les retards et les salaires plus bas, possiblement en raison de facteurs comme :
Moins d'engagement ou de motivation chez les employés moins bien rémunérés.
Des rôles nécessitant moins de supervision, ce qui pourrait permettre plus de retards.

```python
# Boxplot : Retards par département
plt.figure(figsize=(12, 6))
sns.boxplot(x='Department', y='DaysLateLast30', data=data, palette="Set2")
plt.title("Retards par département")
plt.xlabel("Département")
plt.ylabel("Nombre de jours de retard")
plt.xticks(rotation=45, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Les retards sont plus fréquents dans les départements Production, Sales, et Software Engineering, où les employés peuvent faire face à des charges de travail élevées ou à des environnements plus exigeants.
Les départements comme Admin Offices ou Executive Office montrent une meilleure régularité, probablement en raison de la nature administrative et managériale des rôles.

```python
from datetime import datetime

# Calculer la durée d'emploi
data['DateofHire'] = pd.to_datetime(data['DateofHire'])
data['DateofTermination'] = pd.to_datetime(data['DateofTermination'], errors='coerce')
today = datetime.today()

data['Duration'] = (data['DateofTermination'].fillna(today) - data['DateofHire']).dt.days / 365

# Boxplot : Retards et durée d'emploi
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['DaysLateLast30'] > 0, y=data['Duration'], palette="muted")
plt.title("Relation entre les retards et la durée d'emploi")
plt.xlabel("Employés avec retard (1=Oui, 0=Non)")
plt.ylabel("Durée d'emploi (en années)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Durée plus courte pour les employés retardataires :
- Les retards fréquents pourraient être un indicateur de - faible satisfaction ou d'adéquation au poste, menant à des durées d'emploi plus courtes.

Durée plus longue pour les employés sans retard :
- Cela pourrait refléter une plus grande stabilité ou satisfaction dans leur rôle, conduisant à des durées d'emploi prolongées.
